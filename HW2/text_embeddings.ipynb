```python
import numpy as np
from transformers import AutoTokenizer, AutoModel
import torch 
```

```python
layer = 12
# size_embedding = 768 # this is specific to the model you choose

# Load model and tokenizer
model_name = "bert-base-uncased"  # Change to your desired model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

embedding_layer = model.get_input_embeddings()
```

```python
from text_embedding_utils import get_single_string_embedding, get_multi_string_embedding
```

```python
#Examples
print(get_single_string_embedding("heyo", tokenizer, embedding_layer).shape)

#Objects whose names are have multiple words should still be passed as single string
print(get_single_string_embedding("There was an old whore pulling tricks", tokenizer, embedding_layer).shape)
print(get_single_string_embedding("Who at one time could handle five pricks", tokenizer, embedding_layer).shape)
print(get_single_string_embedding("One day she did cry", tokenizer, embedding_layer).shape)
print(get_single_string_embedding("As she pulled out her glass eye,", tokenizer, embedding_layer).shape)
print(get_single_string_embedding("Tell the boys I now can take six!", tokenizer, embedding_layer).shape)

#Multiple strings (actually, just two at a time)
print(get_multi_string_embedding(["This is the same", "but putting it in list"], tokenizer, embedding_layer).shape)
```
